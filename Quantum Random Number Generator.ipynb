```python
# Quantum Random Number Generator
from qiskit import QuantumCircuit
from qiskit_aer import Aer

shots = 1000
qc = QuantumCircuit(1,1)
qc.h(0)          # create 50/50 superposition
qc.measure(0,0)

sim = Aer.get_backend("aer_simulator")
result = sim.run(qc, shots=shots).result()
counts = result.get_counts()
print("counts:", counts)

# produce a random bitstring of length shots (optional)
bits = ''.join(['1' if i=='1' else '0' for i,c in counts.items() for _ in range(c)])[:shots]
print("example bits (first 100):", bits[:100])
```
