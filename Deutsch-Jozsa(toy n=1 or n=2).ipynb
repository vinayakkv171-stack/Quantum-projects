```python
# Deutsch-Jozsa (n=2 example): build a balanced oracle f(x)=x0 XOR x1
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# n = 2 input qubits + 1 ancilla
qc = QuantumCircuit(3,2)

# prepare ancilla in |1> then H
qc.x(2)
qc.h(2)
# Hadamard on inputs
qc.h([0,1])

# Balanced oracle for f(x)=x0 XOR x1: implement as CNOTs from inputs to ancilla
qc.cx(0,2)
qc.cx(1,2)

# Hadamard on inputs again and measure
qc.h([0,1])
qc.measure([0,1],[0,1])

sim = Aer.get_backend("aer_simulator")
res = sim.run(qc, shots=1024).result()
counts = res.get_counts()
print("counts:", counts)
plot_histogram(counts)
plt.show()
# If output is '00' => constant; otherwise balanced.
```
